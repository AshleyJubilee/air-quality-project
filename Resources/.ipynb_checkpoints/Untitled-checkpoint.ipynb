```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('aqi_data.csv')
```

```python
y = data['aqi_category']
x = data['aqi_value']
```

```python
plt.figure(figsize=(30, 15))
plt.scatter(x, y, color='red', marker='o', s=100, alpha=0.7, edgecolor='black', linewidth=0.5)  
plt.title('Scatter Plot')
plt.ylabel('aqi_category')
plt.xlabel('aqi_value')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['aqi_value'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Column')
plt.xlabel('aqi_value')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.plot(data['aqi_category'], data['aqi_value'], marker='o', linestyle='-', color='green')
plt.title('Line Plot')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data['aqi_category'], data['aqi_value'], color='orange')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=89)  
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['aqi_value'], vert=False)
plt.title('Box Plot')
plt.xlabel('Values')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(data['co_aqi_value'], labels=data['aqi_value'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(4, 6))
plt.bar(data['co_aqi_value'], data['aqi_value'], color='orange')
plt.title('Bar Plot')
plt.xlabel('co_aqi_value')
plt.ylabel('aqi_value')
plt.xticks(rotation=89)  
plt.grid(True)
plt.show()
```
